# Calculando a Acurcia do Modelo
# Dados e Rtulos

```python
# dados de treinos x
treino_x = [papagaio1, papagaio2, papagaio3, galinha1, galinha2, galinha3]
# rtulos (classes) treino y
treino_y = [1,1,1,0,0,0]
```

```python
# animais misteriosos - para teste
misterio1 = [1,1,1]
misterio2 = [1,1,0]
misterio3 = [0,1,1]
```

```python
# previses - dados de testes
teste_x = [misterio1, misterio2, misterio3]
# prevendo os testes
previsoes = modelo.predict(teste_x)
```

```python
# rtulos (classes) de teste(avaliao - aqui  a sua provinha)
teste_y = [0,1,1]
```

```python
import sklearn.metrics
```

```python
# importando mdulo que j traz a funo acurcia encapsulada
from sklearn.metrics import accuracy_score

# definindo uma varivel para receber o retorno da funo acurcia * 100 porcetagem
taxa_de_certo = accuracy_score(teste_y, previsoes) * 100

# mostrando acurcia
print(f"Acurcia: {taxa_de_certo:.2f}%")
```
