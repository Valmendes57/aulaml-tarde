```python
! pip install scikit-learn
```

```python
from sklearn.svm import LinearSVC
```

```python

modelo = LinearSVC()
```

```python
modelo.fit(dados, classes)
```

# Prever um animal
# features [1 sim , 0 não]
# tem bico?
# risco de extinção?
# faz cocoricó?

# tem bico, está em extinção e não faz cocoricó



```python
animal_misterioso = [1,1,0]
previsao = modelo.predict([animal_misterioso])
```

```python
previsao
```

# Calculando a Acurácia do Modelo
# Dados e Rótulos

```python
# dados de treinos x
treino_x = [papagaio1, papagaio2, papagaio3, galinha1, galinha2, galinha3]
# rótulos (classes) treino y
treino_y = [1,1,1,0,0,0]
```

```python
# animais misteriosos - para teste
misterio1 = [1,1,1]
misterio2 = [1,1,0]
misterio3 = [0,1,1]
```

```python
# previsões - dados de testes
teste_x = [misterio1, misterio2, misterio3]
# prevendo os testes
previsoes = modelo.predict(teste_x)
```

```python
# rótulos (classes) de teste(avaliação - aqui é a sua provinha)
teste_y = [0,1,1]
```

```python
import sklearn.metrics
```

```python
# importando módulo que já traz a função acurácia encapsulada
from sklearn.metrics import accuracy_score

# definindo uma variável para receber o retorno da função acurácia * 100 porcetagem
taxa_de_certo = accuracy_score(teste_y, previsoes) * 100

# mostrando acurácia
print(f"Acurácia: {taxa_de_certo:.2f}%")
```
